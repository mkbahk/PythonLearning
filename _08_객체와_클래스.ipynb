## 클래스(Class)



### Book 클래스 정의

* 클래스 이름: Book
* 속성
  + 저자: author
  + 책 이름: name
  + 출판사: publisher
  + 발행일: date



```
class Book(object): #object는 상속을 받는 object
  author = ""
  title = ""
  publisher = ""
  date = ""
###
```

```
book = Book()

book.author = "Moon-Kee Bahk"
book.title = "TCP/IP Networking"
print(book.author)
print(book.title)
```

### Book 클래스 메소드 정의

* 메소드
  + 책 정보 출력: `print_info(self)`
  + `self`가 있어야만 실제로 인스턴스가 사용할 수 있는 메소드로 선언
  + `print_info(self)`에서 `self`는 실제적으로 `book` 인스턴스를 의미
  + 메소드 안에서 속성 값을 사용하지 않을 경우에는 `self` 생략 가능

```
class Book(object): #object는 상속을 받는 object
  author = ""
  title = ""
  publisher = ""
  date = ""

  def print_info(self):
    print("Author: ", self.author)
    print("Title: ", self.title)
  ###
###
```

```
book = Book()

book.author = "Moon-Kee Bahk"
book.title = "TCP/IP Networking"
book.print_info()

book2 = Book()
book2.author = "Kee-Beom Bahk"
book2.title = "Ethernet Switch Networking"
book2.print_info()
```

## 인스턴스 속성와 클래스 속성의 활용

* 인스턴스 속성과 클래스 속성을 목적에 맞도록 나누어서 활용

캡처.PNG


### Book 인스턴스 속성과 클래스 속성

* 인스턴스 속성
  + 저자: author
  + 제목: title
  + 출판사: publisher
  + 발행일: date
* 클래스 속성
  + 수량: count

```
class Book3(object): #object는 상속을 받는 object
  author = ""
  title = ""
  publisher = ""
  date = ""
  count = 0

  def print_info(self):
    print("Author: ", self.author)
    print("Title: ", self.title)
    print("Publisher: ", self.publisher)
    print("Date: ", self.date)
  ###
###
```

```
b3 = Book3()

Book3.count += 1

b3.author = "Sang-Bong Bahk"
b3.title = "Python Programming"
b3.publisher = "Kyobo Bookstore"
b3.date = "2021년 10월 31일"
b3.print_info()
print("Number of Books: ", str(Book3.count))
```
